```python
#Scheduling Algorithms 
def sjf_Calculator(a): 
    schedule=[]
    time=0
    while(len(a)>0 ):
        process=[]
        for i in range(len(a)):
            if(a[i][1]<=time):
                process.append(a[i])
        if(len(process)>0):
            process.sort(key = lambda x: x[2])
            for i in range(process[0][2]):
                schedule.append(process[0][0])
            a.remove(process[0])
            time=time+process[0][2]
        else:
            time=time+1
            schedule.append("-")
    return(schedule)
def srtf_Calculator(a):
    schedule=[]
    time=0
    while(len(a)>0):
        process=[]
        for i in range(len(a)):
            if(a[i][1]<=time):
                process.append(a[i])
        if(len(process)>0):
            process.sort(key = lambda x: x[2])
            schedule.append(process[0][0])
            a.remove(process[0])
            time=time+1
            process[0]=(process[0][0],process[0][1],process[0][2]-1)
            if process[0][2]!=0:
                a.append(process[0])
        else:
            time=time+1
            schedule.append("-")
    return(schedule)
def hrrn_Calculator(a):
    schedule=[]
    time=0.0
    while(len(a)>0):
        process=[]
        for i in range(len(a)):
            if(a[i][1]<=time):
                process.append(a[i])
        RR_array=[]
        for i in range(len(process)):
            RR=(time-process[i][1]+process[i][2])/process[i][2]
            RR_array.append(process[i]+[RR])
        if(len(RR_array)>0):
            RR_array.sort(key = lambda x: x[3], reverse=True)
            RR_array[0].pop()
            a.remove(RR_array[0])
            for i in range(RR_array[0][2]):
                schedule.append(RR_array[0][0])
            time=time+RR_array[0][2]
        else:
            time=time+1
            schedule.append("-")
    return(schedule)
def phrrn_Calculator(a):
    schedule=[]
    time=0.0
    while(len(a)>0):
        process=[]
        for i in range(len(a)):
            if(a[i][1]<=time):
                process.append(a[i])
        RR_array=[]
        for i in range(len(process)):
            RR=(time-process[i][1]+process[i][2])/process[i][2]
            g = float("{0:.2f}".format(RR))
            RR_array.append(process[i]+[g])
        if(len(RR_array)>0):
            RR_array.sort(key = lambda x: x[3], reverse=True)
            RR_array[0].pop()
            a.remove(RR_array[0])
            schedule.append(RR_array[0][0])
            time=time+RR_array[0][2]
            process_new_time=[RR_array[0][0],RR_array[0][1],(RR_array[0][2])-1]
            if process_new_time[2]!=0:
                a.append(process_new_time)
        else:
            time=time+1
            schedule.append("-")
    return schedule
def RR_Calculator(a,quantum):
    queue = []
    schedule=[]
    arrange=[]
    time=0
    for i in a:
        if i[1]<=time:
            queue.append(i)
            arrange.append(i)
    a = [i for i in a if i not in arrange]
    arrange=[]
    while(len(a)>0 or len(queue)>0):
        if(len(queue)>0):
            current=queue.pop(0)
            if(current[2]>=quantum):
                for i in range(quantum):
                    schedule.append(current[0])
                current[2]=current[2]-quantum
                time=time+quantum
                for i in a:
                    if i[1]<=time:
                        queue.append(i)
                        arrange.append(i)
                a = [i for i in a if i not in arrange]
                arrange=[]
                if(current[2]!=0):
                    queue.append(current)
            else:
                time=time+current[2]
                for i in range(current[2]):
                    schedule.append(current[0])
                for i in a:
                    if i[1]<=time:
                        queue.append(i)
                        arrange.append(i)
                a = [i for i in a if i not in arrange]
                arrange=[]
        else:
            time=time+1
            schedule.append("-")
            for i in a:
                if i[1]<=time:
                    queue.append(i)
                    arrange.append(i)
            a = [i for i in a if i not in arrange]
            arrange=[]
            
    return(schedule)
def cfs_Calculator(a,q):
    queue = []
    schedule=[]
    arrange=[]
    time=0
    for i in a:
        if i[1]<=time:
            queue.append(i)
            arrange.append(i)
    a = [i for i in a if i not in arrange]
    arrange=[]
    while(len(a)>0 or len(queue)>0):
        if(len(queue)>0):
            quantum=int(q/len(queue))
            if(quantum<=0):
                quantum=1
            current=queue.pop(0)
            if(current[2]>=quantum):
                for i in range(quantum):
                    schedule.append(current[0])
                current[2]=current[2]-quantum
                time=time+quantum
                for i in a:
                    if i[1]<=time:
                        queue.append(i)
                        arrange.append(i)
                a = [i for i in a if i not in arrange]
                arrange=[]
                if(current[2]!=0):
                    queue.append(current)
            else:
                time=time+current[2]
                for i in range(current[2]):
                    schedule.append(current[0])
                for i in a:
                    if i[1]<=time:
                        queue.append(i)
                        arrange.append(i)
                a = [i for i in a if i not in arrange]
                arrange=[]
        else:
            time=time+1
            schedule.append("-")
            for i in a:
                if i[1]<=time:
                    queue.append(i)
                    arrange.append(i)
            a = [i for i in a if i not in arrange]
            arrange=[]
    return(schedule)
def mlfq_Calculator(a,quantum1,quantum2):
    queue1 = []
    queue2 = []
    queue3 = []
    schedule=[]
    time=0
    while(True):
        while(len(queue1)<=0):
            for i in a:
                if i[1]<=time:
                    queue1.append(i)
                    a.remove(i)
            if(len(queue1)==0):
                schedule.append("-")
                time=time+1
        while(len(queue1)>0):
            for i in a:
                    if i[1]<=time:
                        queue1.append(i)
                        a.remove(i)
            current=queue1.pop(0)
            if(current[2]>=quantum1):
                for i in range(quantum1):
                    schedule.append(current[0])
                current[2]=current[2]-quantum1
                time=time+quantum1
                if(current[2]!=0):
                    queue2.append(current)
            else:
                time=time+current[2]
                for i in range(current[2]):
                    schedule.append(current[0])
                
        while(len(queue2)>0):
            for i in a:
                if i[1]<=time:
                    queue1.append(i)
                    a.remove(i)
            if(len(queue1)>0):
                break
            else:
                current=queue2.pop(0)
                if(current[2]>=quantum2):
                    for i in range(quantum2):
                        schedule.append(current[0])
                    current[2]=current[2]-quantum2
                    time=time+quantum2
                    if(current[2]!=0):
                        queue3.append(current)
                else:
                    time=time+current[2]
                    for i in range(current[2]):
                        schedule.append(current[0])
        while(len(queue3)>0):
            for i in a:
                if i[1]<=time:
                    queue1.append(i)
                    a.remove(i)
            if(len(queue1)>0 or len(queue2)>0):
                break
            else:
                queue3.sort(key = lambda x: x[2])
                for i in range(queue3[0][2]):
                    schedule.append(queue3[0][0])
                time=time+queue3[0][2]
                queue3.remove(queue3[0])
        if(len(a)==0 and len(queue1)==0 and len(queue2)==0 and len(queue3)==0):
            break
    return(schedule)
```

```python
#Helper for Gantt Chart
def help_gantt(a):
    current=0
    schedule=[]
    while(current<len(a)):
        count=1
        for i in range(current+1,len(a)):
            if(a[current]==a[i]):
                count=count+1 
            else:
                break
        schedule.append([a[current],count])
        current=current+count
    return schedule     
```

```python
#Creation of gantt chart
import matplotlib.pyplot as plt # Importing the matplotlib.pyplot
def create_figure(schedule):
    fig, gnt = plt.subplots() # Declaring a figure "gnt"
    gnt.set_ylim(0, 7) # Setting Y-axis limits
    fig.set_figwidth(35)
    fig.set_figheight(5)
    gnt.set_yticks([2.5]) # Setting ticks on y-axis
    color_array=['tab:green','tab:olive','tab:purple','tab:cyan','tab:red','tab:pink','tab:orange','tab:brown','tab:blue','tab:green']
    plt.text(0,1,0,color='black',fontsize=27)
    gnt.set_yticklabels([''])
    def addtext(start,period,process):
            gnt.broken_barh([(start,period)], (2, 2), facecolors =color_array[process])
            header="p"+str(process)
            plt.text(start+(period)/2,4.5,header, ha = 'center',color='black',fontsize=30)
            plt.text(start+period,1,start+period, ha = 'center',color='black',fontsize=27)
    # Declaring a bar in schedule
    count=0
    for i in range(len(schedule)):
        if(schedule[i][0]=="-"):
            plt.text(count+schedule[i][1],1,count+schedule[i][1], ha = 'center',color='black',fontsize=27)
            count=count+schedule[i][1]
        else:
            addtext(count,schedule[i][1],schedule[i][0])
            count=count+schedule[i][1]
    plt.savefig('gantt.png')
```

```python
#Functions for waiting time, completion time, turnaround time
def find_statistics(arr,process) :
    x=process[0]
    arrival_time=process[1]
    burst_time=process[2]
    n=len(arr)
    first = -1
    last = -1
    for i in range(0, n) :
        if (x != arr[i]) :
            continue
        if (first == -1) :
            first = i
        last = i 
    if (first != -1) :
        first,last
        turnaround_time=last+1-arrival_time
        waiting_time=turnaround_time-burst_time
        completion=last+1
        return abs(completion),abs(waiting_time),abs(turnaround_time)
    else :
        print("Not Found")
def create_result_array(schedule,process):
    process= [0 if i=="-" else i for i in process]
    process.sort(key=lambda x:x[0])
    result_array=[]
    total_wt=0
    total_tat=0
    for i in range(len(process)):
        if(process[i][0]!=0):
            completion,waiting_time,turnaround_time=find_statistics(schedule,process[i])
            total_wt=total_wt+waiting_time
            total_tat=total_tat+turnaround_time
            result_array.append([process[i][0],completion,waiting_time,turnaround_time])
    total_wt=total_wt/len(process)
    total_wt=f"{total_wt:.4f}"
    total_tat=total_tat/len(process)
    total_tat=f"{total_tat:.4f}"
    return(total_wt,total_tat,result_array)
```
